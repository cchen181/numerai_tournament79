Numerai releases 2 datasets to the user:  numerai_training_data.csv and numerai_tournament_data.csv

These .csvs are too large to post. As a workaround, this script enables others to see *samples* of both datasets.

**Note: The training data is used to create, produce, and validate my model. Following validation, the model is run on the tournament data. Details in ___.ipynb**

```python
##training data
#import libraries
import pandas as pd
import numpy as np

# read data
train = pd.read_csv("./numerai_training_data.csv")
print('TRAINING DATA')
print('Dimensions:')
print(train.shape)
print('Number of rows:')
print(train.shape[0])
print('Number of columns:')
print(train.shape[1])
print('First five rows of training data:')
train.head()
```

```python
##tournament data
# read data
tour = pd.read_csv("./numerai_tournament_data.csv")
print('TOURNAMENT DATA')
print('Dimensions:')
print(tour.shape)
print('Number of rows:')
print(tour.shape[0])
print('Number of columns:')
print(tour.shape[1])
print('First five rows of tournament data:')
tour.head()
```

```python
percent = (train.shape[0]/(train.shape[0]+tour.shape[0]))*100
print('TOTAL DATA')
print('% of total data, training:') #total = training + tournament
print(percent)
print('% of total data, tournament:')
print(100-percent)
```
